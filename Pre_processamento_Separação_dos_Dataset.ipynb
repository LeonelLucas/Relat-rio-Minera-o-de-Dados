```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# **Capturando os dados do dataset original**

```python
df = pd.read_csv("Obfuscated-MalMem2022.csv")
```

```python
SEED = 42
df_attacks = df.iloc[35388:45410].copy()

# Renomear categorias
df_attacks.loc[35388:37389, 'Category'] = '180Solution'
df_attacks.loc[37390:39389, 'Category'] = 'CWS'
df_attacks.loc[39390:41589, 'Category'] = 'Gator'
df_attacks.loc[41590:42999, 'Category'] = 'TIBS'
df_attacks.loc[43000:45409, 'Category'] = 'Transponder'

attack_counts = df_attacks['Category'].value_counts()

print("Amostras por ataque:")
print(attack_counts)

# Média do número de ataques
target_benign = int(attack_counts.mean())

print(f"\nNúmero alvo de benignos: {target_benign}")

df_benign = df[df['Category'] == 'Benign'].sample(
    n=target_benign,
    random_state=SEED
)

df_final = pd.concat([df_attacks, df_benign])

df_final = df_final.sample(frac=1, random_state=SEED).reset_index(drop=True)

# Remover coluna Class
if 'Class' in df_final.columns:
    df_final.drop(columns=['Class'], inplace=True)


df_final.to_csv("dataset_spyware.csv", index=False)

# Verificação final
print("\nDistribuição final:")
print(df_final['Category'].value_counts())
```

## **Dataset Binário**

```python
SEED = 42


df_attacks = df.iloc[35388:45410].copy()
df_attacks['Category'] = 1
n_malicious = len(df_attacks)
print(f"Número total de amostras maliciosas: {n_malicious}")

df_benign = df[df['Category'] == 'Benign'].sample(
    n=n_malicious,
    random_state=SEED
).copy()

df_benign['Category'] = 0
print(f"Número total de benignos: {len(df_benign)}")

df_final = pd.concat([df_attacks, df_benign])

df_final = df_final.sample(
    frac=1,
    random_state=SEED
).reset_index(drop=True)

if 'Class' in df_final.columns:
    df_final.drop(columns=['Class'], inplace=True)

df_final.to_csv("dataset_spyware_binary.csv", index=False)

print("\nDistribuição final:")
print(df_final['Category'].value_counts())
```

```python
dfSpyware = pd.read_csv("dataset_spyware_binary.csv")
```

```python
numeric_cols = dfSpyware.select_dtypes(
    include=['int64', 'float64']
).columns.drop('Category', errors='ignore')

dfSpyware[numeric_cols] = dfSpyware[numeric_cols].astype(np.float32)
dfSpyware['Category'] = dfSpyware['Category'].astype(np.int8)
print(dfSpyware.dtypes)
dfSpyware.to_csv("dataset_spyware_binary_float32.csv", index=False)
```
